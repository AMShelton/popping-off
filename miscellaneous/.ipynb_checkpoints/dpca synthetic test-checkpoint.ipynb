```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_neurons = 6
n_times = 50
activity = np.random.randn(n_neurons, n_times) * 0.1
activity[:3, 30:40] += np.abs(np.random.randn(3, 10))
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
# plot_activity = (activity - np.mean(activity, 0)[np.newaxis, :])
# plot_activity = (activity - np.mean(activity, 1)[:, np.newaxis])
plot_activity = activity.copy()
for ii in range(n_neurons):
    plt.subplot(2, 3, ii + 1)
    if ii < 3:
        ccolor = 'blue'
    else: ccolor = 'red'
    plt.plot(plot_activity[ii, :], linewidth=2, color=ccolor)
    plt.ylim([-2, 2])
    plt.suptitle('(artifial) neuron activity of 6 trials')
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
plot_activity = (activity - np.mean(activity, 0)[np.newaxis, :])
# plot_activity = (activity - np.mean(activity, 1)[:, np.newaxis])
for ii in range(n_neurons):
    plt.subplot(2, 3, ii + 1)
    if ii < 3:
        ccolor = 'blue'
    else: ccolor = 'red'
    plt.plot(activity[ii, :], linewidth=2, color=ccolor, alpha=0.3)
    plt.plot(plot_activity[ii, :], linewidth=2, color=ccolor)
    plt.ylim([-2, 2])
    plt.suptitle('(artifial) neuron activity of 6 trials, after subtraction of mean activity across trials')
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
plot_activity = activity - np.mean(activity)
for ii in range(n_neurons):
    plt.subplot(2, 3, ii + 1)
    if ii < 3:
        ccolor = 'blue'
    else: ccolor = 'red'
    plt.plot(activity[ii, :], linewidth=2, color=ccolor, alpha=0.3)
    plt.plot(plot_activity[ii, :], linewidth=2, color=ccolor)
    plt.ylim([-2, 2])
    plt.suptitle('(artifial) neuron activity of 6 trials, after subtraction mean of all trials and time points')
```
