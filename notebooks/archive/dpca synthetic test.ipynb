```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_neurons = 6
n_times = 50
activity = np.random.randn(n_neurons, n_times) * 0.1
activity[:3, 30:40] += np.abs(np.random.randn(3, 10))
```

```python
plt.rcParams['figure.figsize'] = (12, 4)
def plot_f(plot_activity, axes, reg_activity=None,
           title_str='(artifial) neuron activity of 6 trials'):
    for ii in range(n_neurons):
        if ii < 3:
            row = 0
        else:
            row = 1
        if ii < 3:
            ccolor = 'blue'
        else: ccolor = 'red'
        axes[row][ii % 3].plot(plot_activity[ii, :], linewidth=2, color=ccolor)
        if reg_activity is not None:
            axes[row][ii % 3].plot(reg_activity[ii, :], linewidth=2, color=ccolor, alpha=0.3)
        axes[row][ii % 3].set_ylim([-2, 2])
    plt.suptitle(title_str)

    axes[0][3].plot(plot_activity[:3, :].mean(0), linestyle='--', color='blue', linewidth=2)
    axes[0][3].set_ylim([-2, 2])

    axes[1][3].plot(plot_activity[3:, :].mean(0), linestyle='--', color='red', linewidth=2)
    axes[1][3].set_ylim([-2, 2])
    print(f'Are two means exactly symmetric? : {(plot_activity[:3, :].mean(0) == plot_activity[3:, :].mean(0)).all()}')
    print(f'{(plot_activity[:3, :].mean(0) + plot_activity[3:, :].mean(0))}')
```

```python

fig, ax = plt.subplots(2, 4)
plot_f(plot_activity=activity, axes=ax)
```

```python
fig1, ax1 = plt.subplots(2, 4)
plot_activity1 = (activity - np.mean(activity, 0)[np.newaxis, :])
# plot_activity1 = (activity - np.mean(activity, 1)[:, np.newaxis])
plot_f(plot_activity=plot_activity1, axes=ax1, reg_activity=activity,
       title_str='(artifial) neuron activity of 6 trials, after subtraction of mean activity across trials')
    
    
```

```python
fig2, ax2 = plt.subplots(2, 4)
plot_activity2 = activity - np.mean(activity)
plot_f(plot_activity=plot_activity2, axes=ax2, reg_activity=activity, 
       title_str='(artifial) neuron activity of 6 trials, after subtraction mean of all trials and time points')
    
    
```
